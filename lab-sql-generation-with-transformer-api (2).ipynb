##Set the Question & Prompt and Tokenize
Feel free to change the schema in the prompt below to your own schema

```python
prompt = """### Task
Generate a SQL query to answer [QUESTION]{question}[/QUESTION]

### Instructions
- If you cannot answer the question with the available database schema, return 'I do not know'
- Remember that revenue is price multiplied by quantity
- Remember that cost is supply_price multiplied by quantity

### Database Schema
This query will run on a database whose schema is represented in this string:
CREATE TABLE products (
  product_id INTEGER PRIMARY KEY, -- Unique ID for each product
  name VARCHAR(50), -- Name of the product
  price DECIMAL(10,2), -- Price of each unit of the product
  quantity INTEGER  -- Current quantity in stock
);

CREATE TABLE customers (
   customer_id INTEGER PRIMARY KEY, -- Unique ID for each customer
   name VARCHAR(50), -- Name of the customer
   address VARCHAR(100) -- Mailing address of the customer
);

CREATE TABLE salespeople (
  salesperson_id INTEGER PRIMARY KEY, -- Unique ID for each salesperson
  name VARCHAR(50), -- Name of the salesperson
  region VARCHAR(50) -- Geographic sales region
);

CREATE TABLE sales (
  sale_id INTEGER PRIMARY KEY, -- Unique ID for each sale
  product_id INTEGER, -- ID of product sold
  customer_id INTEGER,  -- ID of customer who made purchase
  salesperson_id INTEGER, -- ID of salesperson who made the sale
  sale_date DATE, -- Date the sale occurred
  quantity INTEGER -- Quantity of product sold
);

CREATE TABLE product_suppliers (
  supplier_id INTEGER PRIMARY KEY, -- Unique ID for each supplier
  product_id INTEGER, -- Product ID supplied
  supply_price DECIMAL(10,2) -- Unit price charged by supplier
);

-- sales.product_id can be joined with products.product_id
-- sales.customer_id can be joined with customers.customer_id
-- sales.salesperson_id can be joined with salespeople.salesperson_id
-- product_suppliers.product_id can be joined with products.product_id

### Answer
Given the database schema, here is the SQL query that answers [QUESTION]{question}[/QUESTION]
[SQL]
"""
```

##Generate the SQL
This can be excruciatingly slow on a T4 in Colab, and can take 10-20 seconds per query. On faster GPUs, this will take ~1-2 seconds

Ideally, you should use `num_beams`=4 for best results. But because of memory constraints, we will stick to just 1 for now.

```python
import sqlparse

def generate_query(question):
    updated_prompt = prompt.format(question=question)
    inputs = tokenizer(updated_prompt, return_tensors="pt").to("cuda")
    generated_ids = model.generate(
        **inputs,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
        max_new_tokens=400,
        do_sample=False,
        num_beams=1,
    )
    outputs = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)

    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    # empty cache so that you do generate more results w/o memory crashing
    # particularly important on Colab â€“ memory management is much more straightforward
    # when running on an inference service
    return sqlparse.format(outputs[0].split("[SQL]")[-1], reindent=True)
```

```python
question = "What was our revenue by product in the New York region last month?"
generated_sql = generate_query(question)
```

```python

import torch
import sqlparse
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
prompt = """### Task
You are a helpful assistant that writes SQL queries.
Generate a SQL query to correctly answer the following question:
[QUESTION]{question}[/QUESTION]

### Guidelines
- Use the database schema provided below.
- If the question cannot be answered using the given schema, return: 'I do not know'.
- Revenue = price * quantity
- Cost = supply_price * quantity
- Always include necessary JOINs based on the foreign keys.

### Database Schema
CREATE TABLE products (
  product_id INTEGER PRIMARY KEY,
  name VARCHAR(50),
  price DECIMAL(10,2),
  quantity INTEGER
);

CREATE TABLE customers (
  customer_id INTEGER PRIMARY KEY,
  name VARCHAR(50),
  address VARCHAR(100)
);

CREATE TABLE salespeople (
  salesperson_id INTEGER PRIMARY KEY,
  name VARCHAR(50),
  region VARCHAR(50)
);

CREATE TABLE sales (
  sale_id INTEGER PRIMARY KEY,
  product_id INTEGER,
  customer_id INTEGER,
  salesperson_id INTEGER,
  sale_date DATE,
  quantity INTEGER
);

CREATE TABLE product_suppliers (
  supplier_id INTEGER PRIMARY KEY,
  product_id INTEGER,
  supply_price DECIMAL(10,2)
);

### Answer
Here is the SQL query that answers [QUESTION]{question}[/QUESTION]:
[SQL]
"""
```

```python
model_name = "microsoft/Phi-3-mini-4k-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16).to("cuda")
```

```python
def generate_query(question):
    updated_prompt = prompt.format(question=question)
    inputs = tokenizer(updated_prompt, return_tensors="pt").to("cuda")

    generated_ids = model.generate(
        **inputs,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
        max_new_tokens=400,
        do_sample=False,
        num_beams=1,
    )

    outputs = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    raw_output = outputs[0]

    # Safely extract the SQL section
    if "[SQL]" in raw_output:
        sql_text = raw_output.split("[SQL]")[-1]
    else:
        sql_text = raw_output

    # Clean and format SQL
    formatted_sql = sqlparse.format(sql_text.strip(), reindent=True, keyword_case="upper")

    # Free GPU memory
    torch.cuda.empty_cache()
    torch.cuda.synchronize()

    return formatted_sql

# ---------------------------------------------
# 4ï¸âƒ£ Example query
# ---------------------------------------------
question = "What was our revenue by product in the New York region last month?"
generated_sql = generate_query(question)

print("\nðŸ§¾ Generated SQL Query:\n")
print(generated_sql)
```

```python
#We learned that the way a prompt is structured has a huge impact on the quality of the modelâ€™s output.
#   A well-organized, explicit prompt helps the model produce accurate, complete SQL queries.
#
# - By improving the promptâ€™s clarity and adding headers like â€œTaskâ€, â€œGuidelinesâ€, â€œDatabase Schemaâ€,
#   and â€œAnswerâ€, the model can follow logical sections and generate results more consistently.
#
# - Giving the model a defined role (â€œYou are a helpful assistant that writes SQL queriesâ€)
#   improves tone and reliability, reducing random or incorrect responses.
#
# - Including explicit rules (e.g., revenue = price * quantity, cost = supply_price * quantity)
#   ensures the model understands business logic and avoids mistakes.
```
